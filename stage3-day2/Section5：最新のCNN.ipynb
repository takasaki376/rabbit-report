#### ゼロから作るディープラーニング　7.7.1　LeNet

LeNetは手書き数字認識を行うネットワークとして1998年に提案されました。畳み込み層とプーリング層を連続して行い、最後に全結合層を経て出力されます。 \
LeNetと「現在のCNN」を比較すると、いくるかの違いがあります。一つ目の違いは活性化関数にあります。LeNetではシグモイド関数が使用されているのに対して、現在では主にReLUが使われます。また、オリジナルのLeNetでは、サブサンプリングによって中間データのサイズ縮小を行っていますが、現在ではMaxプーリングが主流です。

#### ゼロから作るディープラーニング　7.7.2　AlexNet

LeNetが世に出てから２０年近くが経過して、AlexNetが発表されました。このAlexNetがディープラーニングの火付け役になったのですが、ネットワーク構成は基本的にLeNetと大きくは変わっていません。 \
AlexNetは畳み込み層とプーリング層を重ねて、最後に全結合層を経由して結果を出力します。LeNetと大きな構造は変わりませんが、AlexNetでは以下の点が異なります。
- 活性化関数にReLUを用いる
- LRNという局所的正規化を行う層を用いる
- Dropoutを使用する。
