# 第１章：線形代数

線形代数とは、線形という言葉から幾何的な要素と代数的な要素から成り立つ。  
幾何的とは、まっすぐなものを扱う。  
代数的とは、連立方程式をの解や性質を扱う。  

## 代数の成り立ち

連立方程式の解を求める際に、係数を行列とすると、行列の中に解や性質が詰まっている。  
ベクトルの変換を行う。

## 演算規則

### 和
同じ次元のベクトルに対して足し算を行う。

$$ \begin{aligned}
  \left(
    \begin{array}{c}
      ｘ_1 \\
      x_2 \\
      \vdots \\
      x_n
    \end{array}
  \right) 
  +
  \left(
    \begin{array}{c}
      y_1 \\
      y_2 \\
      \vdots \\
      y_n
    \end{array}
  \right) 
  =
  \left(
    \begin{array}{c}
       x_1y_1 \\
       x_2y_2 \\
      \vdots \\
       x_ny_n
    \end{array}
  \right) \\
\end{aligned}$$

### スカラー倍
任意の定数$c$に対して、ベクトルをかける

$$ \begin{aligned}
　　c
  \left(
    \begin{array}{c}
      ｘ_1 \\
      x_2 \\
      \vdots \\
      x_n
    \end{array}
  \right) 
  =
  \left(
    \begin{array}{c}
      cx_1 \\
      cx_2 \\
      \vdots \\
      cx_n
    \end{array}
  \right) \\
\end{aligned}$$

## 固有値、固有ベクトル

ある行列 A に対して，以下のような式が成り立つような，特殊なベクトル $\vec{x}$ と，右辺の係数 $\lambda$ がある。

$$
A\vec{x} = \lambda\vec{x}
$$

行列 A とその特殊なベクトル $\vec{x}$ の積は，ただのスカラーの数 $\lambda$ とその特殊なベクトル $\vec{x}$ との積と同じ値になる。
この特殊なベクト$\vec{x}$ とその係数 $\lambda$ を，行列 A に対する，固有ベクトル，固有値という。

例

$A = \left(
    \begin{array}{cc}
       1 & 4  \\
       2 & 3  \\
    \end{array}
  \right)$ の固有値，固有ベクトルを求めよ

### 固有値の求め方

固有値：$\lambda$、固有ベクトル：$\vec{x}$を求める

\begin{eqnarray}  
    A \vec{x} &= \lambda \vec{x} \\[8pt]
    (A - \lambda I)\vec{x} &= \vec{0}
\end{eqnarray}

$\vec{x} \neq \vec{0}$より

\begin{eqnarray}
   |A - \lambda I| &= 0 \\[8pt]
   \left|
        \begin{array}{cc}
            1 & 4 \\
            2 & 3 \\
        \end{array}
    \right|
    - \lambda 
    \left|
        \begin{array}{cc}
        1 & 0 \\
        0 & 1 \\
        \end{array}
    \right|
    &= 0 \\[8pt]
    \left|
        \begin{array}{cc}
            1 - \lambda & 4 \\
            2 & 3 - \lambda \\
        \end{array}
    \right|
    &= 0 \\[8pt]
    (1 - \lambda)(3 - \lambda) - 4 \times 2 &= 0 \\[8pt]
    (\lambda^2 - 4 \lambda + 3) -8 &= 0 \\[8pt]
    \lambda^2 - 4 \lambda -5 &= 0 \\[8pt]
    (\lambda + 1)(\lambda - 5) &= 0
\end{eqnarray}

つまり、固有値$\lambda = -1 \, or \, 5$ となる。



### 固有ベクトルを求め方

### $\lambda = 5$ の時、$(A - \lambda I)\vec{x} = \vec{0}$ から

\begin{aligned}
    A \vec{x} &= \lambda \vec{x} \\[8pt]
    \left(
        \begin{array}{cc}
            1 & 4 \\
            2 & 3 \\
        \end{array}
    \right)
    \left(
        \begin{array}{c}
            x_1 \\
            x_2 \\
        \end{array}
    \right)
    &= 5 
     \left(
        \begin{array}{c}
            x_1 \\
            x_2 \\
        \end{array}
    \right) \\[12pt]
\end{aligned}
    
$\begin{cases}
            x_1 + 4 x_2 = 5 x_1 \\[4pt]
            2 x_1 + 3 x_2 = 5 x_2 \\
    \end{cases} 
$となり、
$\begin{cases}
            -4 x_1 + 4 x_2 = 0 \\[4pt]
            2 x_1 + -2 x_2 = 0 \\
    \end{cases} 
$　→
$x_1 = x_2$が求まる。


### $\lambda = -1$ の時、$(A - \lambda I)\vec{x} = \vec{0}$ から

\begin{aligned}
    A \vec{x} &= \lambda \vec{x} \\[8pt]
    \left(
        \begin{array}{cc}
            1 & 4 \\
            2 & 3 \\
        \end{array}
    \right)
    \left(
        \begin{array}{c}
            x_1 \\
            x_2 \\
        \end{array}
    \right)
    &= -1 
     \left(
        \begin{array}{c}
            x_1 \\
            x_2 \\
        \end{array}
    \right) \\[12pt]
\end{aligned}

$\begin{cases}
            x_1 + 4 x_2 = -1 x_1 \\[4pt]
            2 x_1 + 3 x_2 = -1 x_2 \\
    \end{cases} 
$となり、
$\begin{cases}
            2 x_1 + 4 x_2 = 0 \\[4pt]
            2 x_1 + 4 x_2 = 0 \\
    \end{cases} 
$　→
$x_1 = -2x_2$が求まる。


### 固有値分解

ある実数を正方形にならべて作られた行列 A が固有値 $\lambda_1$ , $\lambda_2$  ,･･･と固有ベクト  $\vec{v}_1$   $\vec{v}_2$ ,･･･を持ったとする。この固有値を対角線上に並べた行列（それ以外の成分は0）

\begin{aligned}
  \Lambda = \left(
    \begin{array}{c}
      \lambda_1 &  &     \\
       & \lambda_2 &      \\
       &  &   \ddots  \\
    \end{array}
  \right)
\end{aligned}

固有値$\Lambda$の行列に対応する固有ベクトルを並べた行列

\begin{aligned}
V = \left(
    \begin{array}{c}
      v_1 & v_2 & \ldots \\
    \end{array}
  \right)
\end{aligned}

を用意した時、それらは

\begin{aligned}
AV = V \Lambda
\end{aligned}

と関係付けられる。したがって、

\begin{aligned}
A = V \Lambda V^{-1}
\end{aligned}

と変形できる。
このように正方形の行列を上述のような３つの行列の席に変換することを固有値分解という。

固有値分解すると

\begin{aligned}
  \Lambda =
  \left(
    \begin{array}{c}
      ５ & 0   \\
      0 & -1  \\
    \end{array}
  \right) \\
  V = \left(
    \begin{array}{c}
      1 &1  \\
      1 & -\frac{1}{2}   \\
    \end{array}
  \right)
\end{aligned}  

より、

\begin{aligned}  
A &= V \Lambda V^{-1} \\[8px]
  \left(
    \begin{array}{cc}
       1 & 4  \\
       2 & 3  \\
    \end{array}
  \right)
   &= 
  \left(
    \begin{array}{c}
      1 &1  \\
      1 & -\frac{1}{2}   \\
    \end{array}
  \right)
  \left(
    \begin{array}{c}
      ５ & 0   \\
      0 & -1  \\
    \end{array}
  \right)
  \left(
    \begin{array}{c}
      1 &1  \\
      1 & -\frac{1}{2}   \\
    \end{array}
  \right)^{-1}
\end{aligned}

### 利用用途
主成分分析による次元削減を行うときに使用する。　\
ただし、正方行列しか固有値分解はできない。 \
正方行列以外の場合は、次の特異値分解を行う。

## 特異値分解

\begin{aligned}
MV &= US \\[6px]
 M &= USV^{-1}  \\[6px]
 M^TU &= VS^T \\[6px]
 M^T &= VS^TU^{-1}  \\[6px]
\end{aligned}

Uを左特異値ベクトルの行列  
Vを右特異値ベクトルの行列  
Sを特異値の行列

$MM^{-1}$を固有値分解すると、左特異ベクトルと特異値の２乗が求まる。 

\begin{aligned}
MM^{-1} &= (USV^{-1})(VS^TU^{-1}) \\[6px]
 &= USS^TU^{-1} \\[6px]
\end{aligned}

$M^{-1}M$を固有値分解すると、右特異ベクトルと特異値の２乗が求まる。 

\begin{aligned}
M^{-1}M &= (VS^TU^{-1})(USV^{-1}) \\[6px]
 &= VS^TSV^{-1} \\[6px]
\end{aligned}

行列$M$の特異値を２乗すると、$MM^{-1}$、$M^{-1}M$の固有値が求まる。

## 確率

### 頻度確率　（客観確率） 
- 発生する確率 
- 例：「サイコロをふって１が出る確率」 


### ベイズ確率　（主観確率） 
- ある現象の頻度や傾向の代わりに、確率を知識の状態を表す合理的な期待値 
- ある商店街の通りで、通行人が３人連続で男性だった。４人目は男性が通る確率の方が高い 
- 過去の現象から、期待値を求める。たまたま男性が多く通る道なのかも。。。 


### 条件付き確率 
\begin{aligned}
P(A \backslash B) = \frac{P(A \cap B)}{P(B)}
\end{aligned}


### 独立な事象の同時確率 
\begin{aligned}
  P(A \cap B) &= P(A)P(B) \\
  &= P(B \cap A)
\end{aligned}

### ベイズ則
\begin{aligned}
P(A \backslash B) &= \frac{P(A \cap B)}{P(B)} \\[8px]
&= \frac{P(A)P(B \backslash A)}{P(B)}
\end{aligned}
