#### P15

\begin{aligned}
u &= w_1 x_1 + w_2 x_2 + w_3 x_3 + w_4 x_4 + b \\[6px]
&= Wx + b
\end{aligned}

この数式をPythonで書け。

答）

```python
import numpy as np

w1=1
w2=1
w3=1
w4=1

x1=1
x2=1
x3=1
x4=1

b = 1

W = np.array([[w1],[w2],[w3],[w4]])
x = np.array([[x1],[x2],[x3],[x4]])

u = np.dot(W.T ,x) + b
print(u)
```
