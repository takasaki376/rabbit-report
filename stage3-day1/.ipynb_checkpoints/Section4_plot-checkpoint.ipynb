```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd
```

```python
fig = plt.figure()
axes = fig.add_axes((0, 0, 1.5, 1.2), )
axes.set_xlim(-2500,2500) 
axes.set_ylim(-5,130) 
axes.set_xticks([])
axes.set_yticks([])

ber=1000000

x1 = np.linspace(-1000 , 1000)
y1 = (100 * x1 ** 2 ) + 500
plt.plot(x1 , y1/ber ,"black")

x2 = np.linspace(200 , 1000)
y2 = (100000 * x2) - 24999500
plt.plot(x2 , y2/ber ,"red")

x3 = np.linspace(-1000 , -340)
y3 = (-150000 * x3) - 56249500
plt.plot(x3 , y3/ber ,"red")

plt.show
```
